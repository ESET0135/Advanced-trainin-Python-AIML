```python
df=pd.read_csv("Car 1.csv")
```

```python
print(df.isnull().sum())
data_na=df.isnull().sum()*100/df.shape[0]
print(df.isnull().sum()*100/df.shape[0])
```

```python
data_na = data_na.drop(data_na[data_na==0].index).sort_values(ascending=False)
data_na = pd.DataFrame(data_na)
data_na
```

```python
# 1.dropping the Missing Values 
data_dropped=df.dropna()
# data_dropped.isnull().sum()
car_data=df
```

```python
# 2.# 1. groupby Method (fill the value with similer item of value in feature)
car_data['Engine Fuel Type'] = car_data.groupby('Model')['Engine Fuel Type'].transform(lambda x:x.fillna(x.mode().iloc[0] if not x.mode().empty else None))
car_data['Number of Doors'] = car_data.groupby('Model')['Number of Doors'].transform(lambda x:x.fillna(x.mean()))
car_data['Engine Cylinders'] = car_data.groupby('Model')['Engine Cylinders'].transform(lambda x:x.fillna(x.mean()))
car_data['Engine HP'] = car_data.groupby('Model')['Engine HP'].transform(lambda x:x.fillna(x.mean()))

car_data.head(2)
print(car_data.isnull().sum())
# 2. Rule-Based Imputation
print(car_data['Engine Cylinders'].mode(),car_data['Engine HP'].mode())
```

```python
car_data.loc[car_data['Engine Fuel Type'] == 'electric', 'Engine Cylinders'] = 0
car_data[car_data['Engine Cylinders'].isna()]car_data.loc[car_data['Engine Fuel Type'] == 'electric', 'Engine Cylinders'] = 0
```
