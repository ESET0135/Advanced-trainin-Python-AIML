# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns 
```

```python
df=pd.read_csv("museum_visitors.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df[['Firehouse Museum','Avila Adobe','Chinese American Museum','America Tropical Interpretive Center']])
plt.title("All Monthly Users")
plt.xlabel("Date")
plt.ylabel("Visitors")
plt.show()
```

```python
cols=df.columns
for i in cols:
    plt.figure(figsize=(10,6))
    sns.lineplot(data=df[i])
    plt.title(i)
    plt.xlabel("Date")
    plt.ylabel("Visitors")
    plt.show()
```

```python
df3=pd.read_csv('ign_scores.csv',index_col="Platform")
```

```python
df3.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df3['Action'],y=df3.index)
plt.title("AVG rating of action Games by platform")
plt.xlabel("action")
```

# HEatmap

```python
plt.figure(figsize=(8,8))
plt.xlabel('genre')
plt.ylabel('Machine')
plt.title('Gaming ERA')
sns.heatmap(df3,annot=True)
```
