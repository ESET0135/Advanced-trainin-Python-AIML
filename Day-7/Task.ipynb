```python
df.dropna(subset=['Electrical'],inplace=True)
```

```python
# categorical_columns=df[df.columns=='Alley']
categorical_columns=[]
for i in df.columns:
    if (df[i].dtype=='object') and (df[i].isna().sum()>0):
        categorical_columns.append(i)
print(categorical_columns)
    
```

```python
for x in categorical_columns:
    print(x,df[x].value_counts().unique(),df[x].isna().sum())
```

```python
cox=['Alley','MasVnrType','PoolQC','Fence','MiscFeature']
df.drop(columns=cox,inplace=True)
imputation_columns = list(set(categorical_columns) - set(cox))
```

```python
for i in imputation_columns:
    df.fillna(df[i].mode()[0],inplace=True)
```

```python
null_counts = df.isnull().sum()
null_counts = null_counts[null_counts > 0]  # keep only columns that have nulls
print("Columns with missing values and their counts:\n", null_counts)
```

```python
df.info()
```

```python
object_cols = [col for col in df.columns if df[col].dtype == 'object']
print(object_cols)
from sklearn.preprocessing import LabelEncoder

for col in object_cols:
    df[col] = df[col].astype(str)
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
x=df.drop(columns=['SalePrice'])
y=df['SalePrice']
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
LR=LinearRegression(n_jobs=-1)
LR.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_pred=LR.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2_score=r2_score(y_test,y_pred)
```

```python
print(mse,mae,rmse,r2_score*100)
```

```python
from sklearn.metrics import r2_score

# Predictions
y_train_pred = LR.predict(x_train)
y_test_pred = LR.predict(x_test)

# --- Train metrics ---
train_mae = mean_absolute_error(y_train, y_train_pred)
train_mse = mean_squared_error(y_train, y_train_pred)
train_rmse = np.sqrt(train_mse)
train_r2 = r2_score(y_train, y_train_pred)

# --- Test metrics ---
test_mae = mean_absolute_error(y_test, y_test_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_test_pred)

# --- Print results ---
print(f"Train -> MAE: {train_mae:.2f}, RMSE: {train_rmse:.2f}, R2: {train_r2:.2f}")
print(f"Test  -> MAE: {test_mae:.2f}, RMSE: {test_rmse:.2f}, R2: {test_r2:.2f}")
```
