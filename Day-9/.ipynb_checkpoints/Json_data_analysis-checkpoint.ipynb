```python
df_exploded = df.explode(['variants',])
```

```python
df_exploded
```

```python
df_exploded=df_exploded.drop(columns=['stock','reviews','images'])           
```

```python
df_exploded = df_exploded.reset_index()
df_exploded
```

```python
variants_df = pd.json_normalize(df_exploded['variants'])
df_exploded=pd.concat([df_exploded.drop(columns=['variants']),variants_df],axis=1)
df_exploded
```

```python
variants_df = pd.json_normalize(df_exploded['dimensions'])
df_exploded=pd.concat([df_exploded.drop(columns=['dimensions']),variants_df],axis=1)
df_exploded
```

```python
variants_df = pd.json_normalize(df_exploded['ratings'])
df_exploded=pd.concat([df_exploded.drop(columns=['ratings']),variants_df],axis=1)
df_exploded
```
